```python
# import libraries

import serial
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
ser = serial.Serial('COM3', 9600)
```

```python
# ser.close()
```

```python
# remove all serial data from the buffer before reading
ser.reset_input_buffer()
```

```python
data = []

for i in range(50):
    if i ==0:
        continue
    dat = ser.readline()
    data.append(str(dat))
```

```python
t = [float(val.split('. ')[0].split(' ')[1]) for val in data]
h = [float(val.split('. ')[1].split(' ')[1]) for val in data]
c = [float(val.split('. ')[2].split(' ')[1].split('\\')[0]) for val in data]
```

```python
df = pd.DataFrame({'Temperatura':t, 'Humedad':h, 'Conductividad':c})
```

```python
sns.displot(data = df, kde=True, x = 'Temperatura')
plt.show()
```

```python
sns.displot(data = df, kde=True, x = 'Humedad')
plt.show()
```

```python
sns.displot(data = df, kde=True, x = 'Conductividad')
plt.show()
```
